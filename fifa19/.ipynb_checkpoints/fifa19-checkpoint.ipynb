# 3. 선수들의 Overall평균이 높은 상위10개의 팀은 어디일까?

```python
len(set(data['Club']))
```

```python
Clubs = list(set(data['Club']))
```

```python
Clubs[:5]
```

```python
club_mean_overall_list = []

for club in Clubs:
    club_data = data[data.Club == club]
    club_mean_overall = club_data.Overall.mean()
    club_mean_overall_list.append(club_mean_overall)
```

```python
club_mean_overall_list[:5]
```

```python
clubs_mean_overall_df = pd.DataFrame({'Club':Clubs, 'mean_Overall': club_mean_overall_list})
```

```python
clubs_mean_overall_df.head()
```

```python
clubs_mean_overall_df.drop(0).head()
```

```python
clubs_mean_overall_df.sort_values(by='mean_Overall', ascending=False).head(10)
```

```python
Juventus = data[data.Club == 'Juventus']
```

```python
Juventus.Overall.mean()

# 왜 유벤투스가 1등이지..? 하고 확인해보니 호날두가 유벤투스에 있다는 걸 깜빡함...
```

```python
clubs_mean_overall_df.index = clubs_mean_overall_df.Club
```

```python
del clubs_mean_overall_df['Club']
```

```python
clubs_mean_overall_df.sort_values('mean_Overall', ascending=False).head(10)
```

```python
top10 = clubs_mean_overall_df.sort_values('mean_Overall', ascending=False).head(10)
```
